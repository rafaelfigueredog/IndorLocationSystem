```python
import pandas as pd
```

```python
experimento = {}
for i in range(1,16):
    medida = i * 10
    nomeDoAquivo = "dados/experimento1/" + str(medida) + 'cm.txt'
    arquivo = open(nomeDoAquivo, 'r')
    dados = arquivo.read()
    stringSize = len(dados)
    dados = dados[1:stringSize-1]
    dados = dados.split(', ')
    dados = dados[0:100]
    sizeDados = len(dados)
    for i in range(sizeDados):
        dados[i] = int(dados[i])
    coluna = str(medida) + 'cm'
    experimento[coluna] = dados
experimento = pd.DataFrame(data=experimento)
experimento
```
